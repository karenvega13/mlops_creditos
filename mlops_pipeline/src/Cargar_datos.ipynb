**Variable objetivo:** `Pago_atiempo` — indica si el cliente pagó su crédito a tiempo (1) o no (0).


## 1. Importación de librerías

```python
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

print('✅ Librerías importadas correctamente')
print(f'   pandas  : {pd.__version__}')
print(f'   numpy   : {np.__version__}')
```

## 2. Configuración de rutas

```python
# Ruta al dataset (relativa a la raíz del proyecto)
# En producción esta variable apuntaría a la conexión con el DWH
BASE_DIR = os.path.abspath(os.path.join(os.getcwd(), '..', '..'))
DATA_PATH = os.path.join(BASE_DIR, 'Base_de_datos.csv')

print(f'📁 Directorio base   : {BASE_DIR}')
print(f'📄 Ruta del dataset  : {DATA_PATH}')
print(f'🗂️  Archivo existe     : {os.path.exists(DATA_PATH)}')
```

## 3. Carga del dataset

```python
# Carga del archivo CSV
# NOTA: En producción se utilizaría una conexión como:
#   conn = create_engine('postgresql://user:pass@host:port/db')
#   df = pd.read_sql('SELECT * FROM schema.tabla_creditos', conn)

df = pd.read_csv(DATA_PATH, parse_dates=['fecha_prestamo'])

print(f'✅ Dataset cargado exitosamente')
print(f'   Filas    : {df.shape[0]:,}')
print(f'   Columnas : {df.shape[1]}')
```

## 5. Información general del dataset

```python
print('='*60)
print('INFORMACIÓN GENERAL DEL DATASET')
print('='*60)
df.info()
```

```python
# Distribución de la variable objetivo
print('\n📊 Distribución de la variable objetivo (Pago_atiempo):')
print('-'*50)
vc = df['Pago_atiempo'].value_counts()
pct = df['Pago_atiempo'].value_counts(normalize=True) * 100
resumen_target = pd.DataFrame({'Conteo': vc, 'Porcentaje (%)': pct.round(2)})
resumen_target.index = resumen_target.index.map({1: '1 - Pagó a tiempo', 0: '0 - No pagó a tiempo'})
print(resumen_target)
print('\n⚠️  NOTA: El dataset está altamente desbalanceado (~95% / ~5%)')
```

## 6. Validación de carga

```python
# Validaciones básicas para asegurar la calidad de la carga
assert df.shape[0] > 0, 'El dataset no tiene filas'
assert df.shape[1] == 23, f'Se esperaban 23 columnas, se encontraron {df.shape[1]}'
assert 'Pago_atiempo' in df.columns, 'Variable objetivo no encontrada'
assert df['Pago_atiempo'].isin([0, 1]).all(), 'Variable objetivo contiene valores distintos a 0 y 1'

print('✅ Todas las validaciones de carga pasaron exitosamente')
print(f'   Registros cargados: {df.shape[0]:,}')
print(f'   Columnas correctas: {df.shape[1]}')
```
